```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as p
```

```python
df=pd.read_csv("SampleSuperstore.csv")
```

```python
ss=df.drop(['Postal Code'],axis=1)
```

```python
#relationship analysis
```

```python
corelation=ss.corr()
sns.heatmap(corelation,xticklabels=corelation.columns,yticklabels=corelation.columns,annot=True)
```

```python
sns.pairplot(ss)
```

```python
p.figure(figsize=(6,6))
textprops={"fontsize":15}
p.title("Category")
p.pie(df['Category'].value_counts(),labels=df['Category'].value_counts().index,autopct='%1.1f%%',textprops=textprops)
p.show()
```

```python
p.figure(figsize=(10,16))
df.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
p.ylabel('Profit')
p.show()
```

```python
p.figure(figsize=(10,10))
sc=df.groupby(['Sub-Category'])['Profit'].sum().nlargest(30)
sc.plot.barh()
```

```python
p.figure(figsize=(6,6))
p.title("Region")
p.pie(df['Region'].value_counts(),labels=df['Region'].value_counts().index,autopct='%1.1f%%')
p.show()
```

```python
s1=df.groupby("State").Sales.sum().nlargest(n=10)
s2=df.groupby("State").Profit.sum().nlargest(n=10)
s1.plot(kind='bar',figsize=(10,5),fontsize=14)
s2.plot(kind='bar',figsize=(10,5),fontsize=14,color='red')
p.xlabel('states',fontsize=10)
p.ylabel('total sales',fontsize=10)
p.title('sales vs profit',fontsize=10)
p.show()
```
